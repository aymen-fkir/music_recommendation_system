```python
import pandas as pd
import json
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
data = open("output.json",encoding="utf-8")
```

```python
data = json.load(data)
df = pd.DataFrame(data["items"])
```

```python
text = df.song_lyrics
tokenized_output = tokenizer(text.tolist(), return_tensors="pt", padding=True, truncation=True, max_length=512)

with torch.no_grad():
    embeddings = model(**tokenized_output)["last_hidden_state"]
```

```python
df_with_embeddings = pd.DataFrame({
    'text': text,
    'embeddings': embeddings.tolist()  # Convert embeddings to a list for DataFrame
})
```
